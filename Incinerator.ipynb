# Model1 -Here we will be considering the fact that being near or farther to the incinerator will impact on the house prices. Hence in this model we will find out how much will be impact of only 'nearinc' on the house price.

Building linear regression model

```python
from sklearn.model_selection import train_test_split
X = dataset.nearinc.values
X=X.reshape(-1,1)
y = dataset.lprice.values
y=y.reshape(-1,1)
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.8, random_state = 0)
```

```python
from sklearn.linear_model import LinearRegression
linear=LinearRegression()
linear.fit(X_train,y_train)
```

```python
from sklearn import metrics
metrics.r2_score(y_test,linear.predict(X_test))
```

```python
linear.coef_
```

```python
linear.intercept_
```

# Assumption & Solution



From the given coefficient of the above linear model, it is clear that if the house is near the dumpsite (i.e within 15600ft) then price drop will be around 3.9% for every 1ft of closeness. so, we can say that the house which is near to the Incinerator have low price and the house which is far have the higher price.

# Model2

In this model we are finding the impact of age,land,area and incinerator distance from the house on the selling price..Hence in this model  we will find out how much will be impact of all the above features on the house price.

```python
feat=['larea','lland','ldist','age','nearinc','agesq']
X = dataset[feat]
y = dataset[['lprice']]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.8, random_state = 0)
```

```python
linear=LinearRegression()
linear.fit(X_train,y_train)
```

```python
metrics.r2_score(y_test,linear.predict(X_test))
```

```python
features=[]
for x in feat:
  features.append(x)
coeff=[]
for y in linear.coef_[0]:
    coeff.append(y)

res = {} 
for key in features: 
    for value in coeff: 
        res[key] = value 
        coeff.remove(value) 
        break  
```

```python
res
```

# Assumption & Solution

 From the above model, we have analysed that: 1) For increase of area by 1% there will be a price increase by 63%,assuming all other variables are held constant. 2)For increase of land by 1% there will be a price increase by 0.4%,assuming all other variables are held constant.3) for increase of dist. from house to incinerator, feet pricecrease of 12%,assuming all other variables are held constant. 4) For each year increase in house age, there will be 1% of decrease in house price,assuming all other variables are held constant.From the given coefficient of the above linear model, it is clear that if the house is near the dumpsite (i.e within 15600ft) then price drop will be around 19.4% for every 1ft of closeness.
